# Data Preprocessing

###Steps and Justifications

Preprocessing is a crucial step in any NLP pipeline. Raw text data is often noisy and inconsistent, which can hurt model performance. Here's a breakdown of the key preprocessing steps that were performed and the rationale behind each:

1. **Removing Special Characters**  
   Filters out punctuation, digits, and special characters using regular expressions (e.g., "nice!!! 100%" → "nice").  
   **Why?** Most ML models don't benefit from symbols unless explicitly modeled; keeping them adds noise.

2. **Lowercasing**  
   Converts all characters to lowercase (e.g., "Great" → "great").  
   **Why?** Helps reduce the vocabulary size by treating "Great" and "great" as the same word.

3. **Tokenization**  
   Splits the text into individual words or tokens (e.g., "great product" → ["great", "product"]).  
   **Why?** This is essential to analyze word-level patterns and apply further filtering.

4. **Stopword Removal**  
   Removes common words like "the", "is", "and", which do not carry meaningful sentiment or deception cues.  
   **Why?** Reduces noise and dimensionality, allowing the model to focus on content-bearing words.

5. **Lemmatization**  
   Converts each word to its base form or lemma (e.g., "running" → "run", "better" → "good").  
   **Why?** Ensures that different inflections of the same word are treated as one, improving generalisation.

<br>

Each of these steps ensures that the text is cleaned, normalised, and optimised for feature extraction using techniques like Bag-of-Words, ultimately improving model accuracy and interpretability.

## Feature Extraction: Bag of Words (BoW)

Bag-of-Words (CountVectorizer) is used to convert text into numeric vectors, which are then used to train the ML model. In this approach, each document (review) is represented as a vector of word counts.

*How it works:*  
- A vocabulary of all unique words in the corpus is created.
- Each review is then encoded as a vector indicating the frequency (or presence) of each word from the vocabulary.

*Why Bag of Words?*  
- It's simple and effective for baseline models.
- It captures how frequently words appear, which can be useful for detecting fake reviews (e.g., overuse of certain marketing terms).

*Limitations:*  
- Ignores word order and context.
- High dimensional and sparse, especially for large corpora.

Despite its limitations, BoW works well with linear models like SVMs and is computationally efficient for medium-sized datasets.


```python
# Train Bag of Words model

from sklearn.feature_extraction.text import CountVectorizer

count_vec = CountVectorizer()

X_train_cv = count_vec.fit_transform(X_train)

X_train_cv.shape
```

## Model Training: Support Vector Machine (SVM)

I used a SVM model with a linear kernel for classifying reviews as fake or real.

*What is SVM?*  
Support Vector Machines are supervised learning algorithms used for binary classification. They work by finding the optimal hyperplane that separates data points of different classes in high-dimensional (in our data - 30925 dimensional) space.

*Why SVM for Fake Review Detection?*  
- **Effective in high-dimensional spaces:** BoW results in high-dimensional feature vectors; SVM handles this well.
- **Robust to overfitting:** Particularly when using a linear kernel with proper regularization.
- **Strong baseline:** Often performs competitively with more complex models when data is linearly separable.

I used `LinearSVC` from Scikit-learn because it's optimised for text classification and large feature sets.


**Regularization in Support Vector Machines:**

Regularization is a crucial concept in Support Vector Machines (SVMs), especially when working with noisy, high-dimensional data like text.

In the context of `LinearSVC` (Scikit-learn's implementation of linear SVM), regularization is controlled by the **`C` parameter**.

---

What is Regularization?

Regularization helps prevent **overfitting** by discouraging the model from fitting too closely to the training data. It balances the trade-off between:

- **Maximizing the margin** (i.e., separating classes with a wide boundary), and  
- **Minimizing classification errors** on the training data.

---

The Role of `C`

- A **small `C`** means more emphasis on a large margin:
  - The model is penalized less heavily for misclassifying training points.
  - It prefers a simpler decision boundary, even if that means more training errors. The goal here is to maximise the margin in between the separating hyperplane and the 2 clusters of points.
- A **large `C`** means more emphasis on reducing training loss:
  - The model focuses more on minimizing training error.
  - It allows more complex decision boundaries, possibly at the risk of overfitting.

---

Why It Matters?

Since we’re using **high-dimensional Bag-of-Words features**, regularization helps the SVM generalize better and avoid overfitting to spurious word patterns that only appear in the training set.

We can tune `C` value via cross-validation to find the best trade-off between training performance and generalization.

```python
# SVM model using linear svc with hinge loss and l2 regularization, regularization parameter 1.0. Show training accuracy

from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

# Train a LinearSVC model with hinge loss and l2 regularization
model = LinearSVC(loss='hinge', penalty='l2', C=1.0)
model.fit(X_train_cv, y_train)

# Calculate the training accuracy
y_pred_train = model.predict(X_train_cv)
train_accuracy = accuracy_score(y_train, y_pred_train)

# Print the training accuracy
print('Training Accuracy:', train_accuracy)
```

###2. SVM Model (5-fold cross validation)
- hinge loss
- l2 regularization
- regularization hyperparameter (C) = 0.15
- 5 fold cross validation

To evaluate the robustness and generalizability of our SVM model, I used **5-fold cross-validation**.

---
**What is Cross-Validation?**

Cross-validation is a resampling technique that helps us estimate how well a model will perform on unseen data.

In **5-fold cross-validation**:
1. Specify a set of different regularization hyperparameter ('C') values that you would like to test
2. The dataset is split into 5 equal parts (folds).

Loop through the set of parameter values, and for each paramter value, do the following:

3. In each iteration, 4 folds are used for training and the remaining 1 for validation.
4. Step 3 is repeated 5 times, with each fold used exactly once as the validation subset.
5. The final performance is the average validation preformance across 5 runs.

SELECT THE REGULARIZATION HYPERPARAMETER VALUE THAT GIVES THE HIGHEST AVERAGE VALIDATION ACCURACY.

---

**Why Use It?**

- **Reduces bias and variance** in evaluation compared to a single train-test split.

- Helps us spot overfitting if model accuracy varies widely across folds.

In this project, 5-fold cross-validation helps specify the regularization hyperparameter such that the


```python
# run a SVM modelusing linear svc with hinge loss and l2 regularization. Use 5 fold cross validation to decide regularization parameter

from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC

# Define the model
model_FRD = LinearSVC(loss='hinge', penalty='l2')

# Define the hyperparameter grid
param_grid = {'C': [0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 1.0, 10.0]}

# Perform grid search with cross-validation
grid_search = GridSearchCV(model_FRD, param_grid, cv=5)
grid_search.fit(X_train_cv, y_train)

# Get the best model
best_model_FRD = grid_search.best_estimator_

# Print the best hyperparameters
print('Best C:', best_model_FRD.C)

# Calculate the training accuracy
y_pred_train = best_model_FRD.predict(X_train_cv)
train_accuracy = accuracy_score(y_train, y_pred_train)

# Print the training accuracy
print('Training Accuracy:', train_accuracy)

# Evaluate the best model on the test set
y_pred_test = best_model_FRD.predict(count_vec.transform(X_test))
test_accuracy = accuracy_score(y_test, y_pred_test)

# Print the test accuracy
print('Test accuracy:', test_accuracy)
```

```python
# confusion matrix and classification report for the above model

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Predict on the test set
y_pred_test = best_model_FRD.predict(count_vec.transform(X_test))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# Classification Report
print(classification_report(y_test, y_pred_test))
```

### Evaluation Metrics

1. **Accuracy**  
   Overall percentage of correctly predicted reviews.

2. **Precision**  
   Of all reviews predicted as fake, how many were actually fake?  
   *Important when false positives are costly (e.g., mislabeling a genuine reviewer).*

3. **Recall**  
   Of all actual fake reviews, how many did we detect?  
   *Important for catching as many fakes as possible.*

4. **F1 Score**  
   Harmonic mean of precision and recall.  
   *Useful when there’s a tradeoff between catching more fake reviews and avoiding false alarms.*

Using these together gives a fuller picture of model performance than accuracy alone.
